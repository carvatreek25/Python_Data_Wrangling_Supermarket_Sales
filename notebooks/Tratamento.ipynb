Parte II do Projeto  - Transformação 
Aqui eu vou mudar os nomes das colunas para português e criar algumas colunas que vão ajudar na análise

```python
# Lendo o arquivo CSV
df = pd.read_csv('data/supermarket_sales.csv')
```

```python
# Aqui eu vou criar um dicionário das colunas em inglês para o português
colunas_traducao = {
        'Invoice ID': 'id_fatura',
        'Branch': 'filial',
        'City': 'cidade',
        'Customer type': 'tipo_cliente',
        'Gender': 'genero',
        'Product line': 'linha_produto',
        'Unit price': 'preco_unitario',
        'Quantity': 'quantidade',
        'Tax 5%': 'imposto_5pct',
        'Total': 'total',
        'Date': 'data',
        'Time': 'hora',
        'Payment': 'forma_pagamento',
        'cogs': 'custo_produtos',
        'gross margin percentage': 'margem_bruta_pct',
        'gross income': 'lucro_bruto',
        'Rating': 'avaliacao'
    }
df.rename(columns=colunas_traducao, inplace=True)
print(f"{colunas_traducao}")
```

```python
# Agora eu vou mudar o nome de alguns valores para português
 # Tipo de Cliente
df['tipo_cliente'] = df['tipo_cliente'].map({
        'Member': 'Membro',
        'Normal': 'Normal'
    })
    
    # Gênero
df['genero'] = df['genero'].map({
        'Male': 'Masculino',
        'Female': 'Feminino'
    })
    
    # Forma de Pagamento
df['forma_pagamento'] = df['forma_pagamento'].map({
        'Ewallet': 'Carteira Digital',
        'Cash': 'Dinheiro',
        'Credit card': 'Cartão de Crédito'
    })
    
    # Linha de Produto
linha_produto_traducao = {
        'Electronic accessories': 'Acessórios Eletrônicos',
        'Fashion accessories': 'Acessórios de Moda',
        'Food and beverages': 'Alimentos e Bebidas',
        'Health and beauty': 'Saúde e Beleza',
        'Home and lifestyle': 'Casa e Estilo de Vida',
        'Sports and travel': 'Esportes e Viagem'
    }
df['linha_produto'] = df['linha_produto'].map(linha_produto_traducao)
    
```

```python
df.head(10)
```

```python
# 1. Conversão inicial (Garante que o Pandas entenda Data e Hora)
df['data'] = pd.to_datetime(df['data'], format='mixed')
temp_time = pd.to_datetime(df['hora'], format='%H:%M')

# 2. Extrações de Data
df['ano'] = df['data'].dt.year
df['mes'] = df['data'].dt.month
df['dia'] = df['data'].dt.day
df['trimestre'] = df['data'].dt.quarter
df['semana_ano'] = df['data'].dt.isocalendar().week

# 3. Tratamento de Hora e Minuto (Posicionamento Lado a Lado)
idx_hora = df.columns.get_loc('hora')
if 'minuto' in df.columns:
df.drop(columns=['minuto'], inplace=True)
df.insert(idx_hora + 1, 'minuto', temp_time.dt.minute)
df['hora'] = temp_time.dt.hour # Agora 'hora' é um número (0-23)

# 4. Mapeamento de Nomes (Meses e Dias da Semana)
df['mes_nome'] = df['data'].dt.month_name().map({
    'January': 'Janeiro', 'February': 'Fevereiro', 'March': 'Março',
    'April': 'Abril', 'May': 'Maio', 'June': 'Junho',
    'July': 'Julho', 'August': 'Agosto', 'September': 'Setembro',
    'October': 'Outubro', 'November': 'Novembro', 'December': 'Dezembro'
})

df['dia_semana'] = df['data'].dt.day_name().map({
    'Monday': 'Segunda-feira', 'Tuesday': 'Terça-feira', 'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira', 'Friday': 'Sexta-feira', 'Saturday': 'Sábado', 'Sunday': 'Domingo'
})

# 5. REGRAS DE NEGÓCIO (O que você pediu para incluir)

# Função para Período do Dia
def classificar_periodo(h):
    if 6 <= h < 12: return 'Manhã'
    elif 12 <= h < 18: return 'Tarde'
    elif 18 <= h < 22: return 'Noite'
    else: return 'Madrugada'

df['periodo_dia'] = df['hora'].apply(classificar_periodo)

# Dia Útil (Usando a data que ainda é datetime)
df['dia_util'] = (df['data'].dt.dayofweek.isin([5, 6])).map({True: 'Sim', False: 'Não'})

# Quinzena (Usando a coluna 'dia' que extraímos acima)
df['quinzena'] = np.where(df['dia'] <= 15, 'Primeira Quinzena', 'Segunda Quinzena')

# Fim de Semana (Binário)
df['eh_fim_semana'] = df['data'].dt.dayofweek.isin([5, 6]).astype(int)

# 6. Formatação Final para Exibição
# Somente agora transformamos a data em texto brasileiro
df['data_visual'] = df['data'].dt.strftime('%d/%m/%Y')

df.head(10)
```

```python
# Agora vou finalizar o arquivo salvando o CSV tratado  
df.to_csv('data/supermarket_sales_tratado.csv', index=False)
```
